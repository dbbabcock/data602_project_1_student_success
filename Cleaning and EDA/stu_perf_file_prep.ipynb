In order to create a larger dataset, I downloaded 10 batches of of the large sample (1,000 rows). In this notebook, I combine those into a single 10,000 row CSV file to use for the rest of the project.

```python
import glob
import pandas as pd
print('Pandas version used: ' + pd.__version__)
```

```python
# Find all of the files in this directory with the .csv format
all_exam_files = glob.glob('../Datasets/separate_exam_files/*.csv')
```

```python
# Make sure we got all of the correct files:
all_exam_files
```

```python
# Read each of the files into a dataframe and then concatenate the dataframes together:
stu_perf_raw = pd.concat(pd.read_csv(file) for file in all_exam_files)
```

```python
# Reset the index so that is goes from 0->9999 instead of 0->99 10 times:
stu_perf_raw.reset_index(drop=True, inplace=True)
```

```python
# Final check before saving the combined file:
stu_perf_raw
```

```python
# Save the new csv file to the Datasets folder:
stu_perf_raw.to_csv("../Datasets/stu_perf_raw.csv", index=False)
```
